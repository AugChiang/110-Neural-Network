```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"history.csv")
df2 = pd.read_csv(r"result.csv")
print("Model B",df.head(5))
print("Model A", df2.head(5))
```

```python
d_loss = df['D-loss']
g_loss = df['G-loss']
d_acc = df['D-acc'].str.rstrip('%').astype('float') / 100.0 # convert percentage to float
# print(d_loss.head(5))
# print(g_loss.head(5))
# d_acc.head(5)
```

```python
d2_loss = df2['d-loss']
g2_loss = df2['g-loss']
d2_acc = df2['d-acc']
# print(d_loss.head(5))
# print(g_loss.head(5))
```

```python
epoch_every100 = np.arange(0,10000,step=100)
d_every100_loss = d_loss[0::100]
d2_every100_loss = d2_loss[0::100]

plt.plot(d2_every100_loss, 'red', label="model A")
plt.plot(d_every100_loss, 'blue', label="model B")
plt.title("Discriminator Loss by Epochs (step=100)")
plt.xlabel("# Epochs")
plt.ylabel("Loss")
plt.legend(["model A", "model B"])
plt.show
```

```python
g_every50_loss = g_loss[0::50]
g2_every50_loss = g2_loss[0::50]

plt.plot(g2_every50_loss, 'red', label="model A")
plt.plot(g_every50_loss, 'blue', label="model B")
plt.title("Generator Loss by Epochs (step=50)")
plt.xlabel("# Epochs")
plt.ylabel("Loss")
plt.legend(["model A", "model B"])
plt.show
```

```python
d_every50_acc = d_acc[0::50]
d2_every50_acc = d2_acc[0::50]

plt.plot(d2_every50_acc, 'red', label="model A")
plt.plot(d_every50_acc, 'blue', label="model B")
plt.title("Discriminator Accuracy by Epochs (step=50)")
plt.xlabel("# Epochs")
plt.ylabel("acc")
plt.legend(["model A", "model B"])
plt.show
```

```python
df100k = pd.read_csv(r"e100k.csv")
df100k.head(5)
```

```python
d100k_loss = df100k['d-loss']
g100k_loss = df100k['g-loss']
d100k_acc = df100k['d-acc'].str.rstrip('%').astype('float') / 100.0 # convert percentage to float
```
